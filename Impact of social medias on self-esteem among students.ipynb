```python
import pandas as pd
```

```python
df1=pd.read_csv('/content/Research Self-esteem (Responses) - Form Responses 1.csv')
df1.info()
```

```python
df1=df1.rename(columns={"On average, how often do you use Facebook":"FB","On average, how often do you use Instagram":"Insta","On average, how often do you use Snapchat":"Snap","On average, how often do you use Youtube":"YouTube","On average, how often do you use LinkedIn":"LinkedIn","On average, how often do you use Reddit":"Reddit","On average, how often do you use X(Formerly Twitter)":"Twitter"},errors="raise")
df1.info()
```

```python
df1=df1.rename(columns={"On the whole, I am satisfied with myself. ":"SE1"})
df1=df1.rename(columns={"At times I think I am no good at all. ":"SE2"})
df1=df1.rename(columns={"I feel that I have a number of good qualities. ":"SE3"})
df1=df1.rename(columns={"I am able to do things as well as most other people. ":"SE4"})
df1=df1.rename(columns={"I feel I do not have much to be proud of. ":"SE5"})
df1=df1.rename(columns={"I certainly feel useless at times. ":"SE6"})
df1=df1.rename(columns={"I feel that I'm a person of worth. ":"SE7"})
df1=df1.rename(columns={"I wish I could have more respect for myself":"SE8"})
df1=df1.rename(columns={"All in all, I am inclined to think that I am a failure. ":"SE9"})
df1=df1.rename(columns={"I take a positive attitude toward myself.":"SE10"})
df1.info()
```

```python
df1=df1.rename(columns={"On average how much time you use your mobile phone each day(in hrs)":"Avg_ScreenTime"})
df1.info()
```

```python
df1.drop("Timestamp",axis=1,inplace=True)
```

```python
df1['Ethnicity']="Indian"
```

```python
apps=['FB','Insta','Snap','YouTube','LinkedIn','Reddit','Twitter']
values={"Multiple times a day":7,"Once a day":6,"A few times a week":5,"Once a week":4,"Once or twice a month":3,"Less than once a month":2,"Not at all":1}
for i in apps:
  df1[i]=df1[i].replace(values)
df1
```

```python
ques=["SE1","SE2","SE4","SE6","SE7"]
values={"Strongly agree":3,"Agree":2,"Neutral":0,"Disagree":1,"Strongly disagree":0}
for i in ques:
  df1[i]=df1[i].replace(values)
df1
```

```python
ques=["SE3","SE5","SE8","SE9","SE10"]
values={"Strongly agree":0,"Agree":1,"Neutral":0,"Disagree":2,"Strongly disagree":3}
for i in ques:
  df1[i]=df1[i].replace(values)
df1
```

```python
object_columns = df1.select_dtypes(include=['object']).columns
print(object_columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,15))
sns.heatmap(df1.corr(),cmap='coolwarm',annot=True)
plt.figure(figsize=(15,15))
plt.show()
```

```python
df_LR=df1
df_LR.drop('FB',axis=1,inplace=True)
df_LR.drop('Insta',axis=1,inplace=True)
df_LR.drop('Snap',axis=1,inplace=True)
df_LR.drop('YouTube',axis=1,inplace=True)
df_LR.drop('LinkedIn',axis=1,inplace=True)
df_LR.drop('Reddit',axis=1,inplace=True)
df_LR.drop('Twitter',axis=1,inplace=True)
```

```python
object_columns = df_LR.select_dtypes(include=['object']).columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in object_columns:
  df_LR[i]=le.fit_transform(df_LR[i])
```

```python
df_LR.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
import math
lr=LinearRegression()
y=df_LR[['Sum']]
df_LR.drop('Sum',axis=1,inplace=True)
x=df_LR
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=2)
lr.fit(xtrain,ytrain)
```

MULTI-LINEAR REGRESSION

```python
ypred=lr.predict(xtest)
mse_lr=mean_squared_error(ytest,ypred)
print("MSE: ",mse_lr)
mae_lr=mean_absolute_error(ytest,ypred)
print("MAE: ",mae_lr)
rmse_lr = math.sqrt(mse_lr)
print("RMSE: ",rmse_lr)
r2 = r2_score(ytest, ypred)
print("R-squared:", r2)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2)
print(lr.coef_)
print(lr.intercept_)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain, ytrain)
y_pred_rf = rf.predict(xtest)
mse_rf = mean_squared_error(ytest, y_pred_rf)
print("Mean Squared Error (MSE):", mse_rf)
mae_rf = mean_absolute_error(ytest, y_pred_rf)
print("Mean Absolute Error (MAE):", mae_rf)
rmse_rf = math.sqrt(mse_rf)
print("Root Mean Squared Error (RMSE):", rmse_rf)
r2_rf = r2_score(ytest, ypred)
print("R-squared:", r2_rf)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2_rf = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2_rf)
```

SUPPORT VECTOR MACHINE

```python
from sklearn.svm import SVR
svm_regressor = SVR(kernel='linear')
svm_regressor.fit(xtrain, ytrain)
y_pred_svm = svm_regressor.predict(xtest)
mse_svm = mean_squared_error(ytest, y_pred_svm)
print("Mean Squared Error (MSE):", mse_svm)
mae_svm = mean_absolute_error(ytest, y_pred_svm)
print("Mean Absolute Error (MAE):", mae_svm)
rmse_svm = math.sqrt(mse_svm)
print("Root Mean Squared Error (RMSE):", rmse_svm)
r2_svm = r2_score(ytest, y_pred_svm)
print("R-squared:", r2_svm)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2_svm = 1 - (1 - r2_svm) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2_svm)
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeRegressor
dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(xtrain, ytrain)
y_pred_dt = dt_regressor.predict(xtest)
mse_dt = mean_squared_error(ytest, y_pred_dt)
print("Mean Squared Error (MSE):", mse_dt)
mae_dt = mean_absolute_error(ytest, y_pred_dt)
print("Mean Absolute Error (MAE):", mae_dt)
rmse_dt = math.sqrt(mse_dt)
print("Root Mean Squared Error (RMSE):", rmse_dt)
r2_dt = r2_score(ytest, y_pred_dt)
print("R-squared:", r2_dt)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2_dt = 1 - (1 - r2_dt) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2_dt)
```

RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge
ridge_regressor = Ridge(alpha=1.0)
ridge_regressor.fit(xtrain, ytrain)
y_pred_ridge = ridge_regressor.predict(xtest)
mse_ridge = mean_squared_error(ytest, y_pred_ridge)
print("Mean Squared Error (MSE):", mse_ridge)
mae_ridge = mean_absolute_error(ytest, y_pred_ridge)
print("Mean Absolute Error (MAE):", mae_ridge)
rmse_ridge = math.sqrt(mse_ridge)
print("Root Mean Squared Error (RMSE):", rmse_ridge)
r2_ridge = r2_score(ytest, y_pred_ridge)
print("R-squared:", r2_ridge)
n = xtest.shape[0]
p = xtest.shape[1]
adjusted_r2_ridge = 1 - (1 - r2_ridge) * (n - 1) / (n - p - 1)
print("Adjusted R-squared:", adjusted_r2_ridge)
```

```python
metrics = {
    'Model': ['Linear Regression', 'Random Forest', 'SVM', 'Decision Tree', 'Ridge Regression'],
    'MSE': [mse_lr, mse_rf, mse_svm, mse_dt, mse_ridge],
    'MAE': [mae_lr, mae_rf, mae_svm, mae_dt, mae_ridge],
    'RMSE': [rmse_lr, rmse_rf, rmse_svm, rmse_dt, rmse_ridge],
    'R-squared': [r2, r2_rf, r2_svm, r2_dt, r2_ridge],
    'Adjusted R-squared': [adjusted_r2, adjusted_r2_rf, adjusted_r2_svm, adjusted_r2_dt, adjusted_r2_ridge]
}
metrics_df = pd.DataFrame(metrics)
metrics_df
```
